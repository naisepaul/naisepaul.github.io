Use gradient descent/ascent on each of the target functions  
o Plot the best performance of this method vs number of points examined (15 marks) 

o Plot the points of the function this method examines. (15 marks) 

```python
def f(x): 
    return (x-1)**2
def f_derivative(x):
    return 2*x-2

x = linspace(-9,11,150)
y = f(x)

a = 6
h = 0.1
n=0
#plt.figure(figsize=(15,5))
while n<10:
    fprime = (f(a+h)-f(a))/h # derivative    
    tan = f(a)+fprime*(x-a)  # tangent
    gradient= f_derivative(a)           
    plt.plot(x,y,'b',a,f(a),'om',x,tan,'--r')#plot of the function and the tangent
    if gradient ==0: break
    else:
        a = a-1
        n+=1 
    
print "slop is zero at :", a   # function is minimum at a=1
```

```python
x_old = 0 
x_new =6
step_size = 0.01 
precision = 0.00001

def f(x):
    return 4*(x**3)-6*(x**2)-4.5*x + 4.5

while abs(x_new - x_old) > precision:
    x_old = x_new
    gradient= f(x_old)
    x_new = x_old - step_size * gradient
    
print("Local minimum occurs at", round(x_new,2))
```
